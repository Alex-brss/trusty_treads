```python
import pandas as pd

sneakersnstuff = pd.read_csv("sneakersnstuff.csv", sep=";")
```

```python
sneakersnstuff = sneakersnstuff[sneakersnstuff["available"] == "InStock"]
del sneakersnstuff["available"]
```

```python
sneakersnstuff.head()
```

```python
sneakersnstuff["brand"] = sneakersnstuff["brand"].str[0].str.upper() + sneakersnstuff["brand"].str[1:]
```

```python
scrapdate = "07-12-2023"
```

```python
# rename url to link
sneakersnstuff.rename(columns={"url": "link"}, inplace=True)

# rename img to image
sneakersnstuff.rename(columns={"img": "image"}, inplace=True)
```

```python
sneakersnstuff["_id"] = range(1, len(sneakersnstuff) + 1)
sneakersnstuff["scrapDate"] = scrapdate
```

```python
sneakersnstuff.head()
```

```python
vinted = pd.read_csv("new_vinted.csv", sep=";")
```

```python
for key, products in vinted.iterrows():
    if products["name"] not in sneakersnstuff["name"].values:
        vinted.drop(key, inplace=True)
```

```python
vinted.drop_duplicates(subset=["brand","name"], keep="first", inplace=True)
```

```python
vinted
```

```python
df = sneakersnstuff.copy()

list_vinted_url = []
list_vinted_price = []
list_vinted_stars = []

for key, products in df.iterrows():
    find = False
    for key2, products2 in vinted.iterrows():
        if  products2["name"] == products["name"]:
            list_vinted_url.append(products2["url"])
            list_vinted_price.append(products2["price"])
            list_vinted_stars.append(3.3)
            find = True
            break
    if find == False:
        list_vinted_url.append("NaN")
        list_vinted_price.append("NaN")
        list_vinted_stars.append("NaN")

df["vinted_url"] = list_vinted_url
df["vinted_price"] = list_vinted_price
df["vinted_stars"] = list_vinted_stars
```
