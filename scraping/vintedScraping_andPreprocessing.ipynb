### Initializing scraper 

```python
import pandas as pd 
import numpy as np
from vinted_scraper import VintedScraper

# We use the VintedScraper library to scrape the website
scraper = VintedScraper("https://www.vinted.fr")

df = pd.read_csv('C:/Users/alexa/Downloads/sneakersnstuff.csv',sep=';')
df.head()

```

```python
# We drop the duplicates
df[['brand','name','color']].drop_duplicates().count()
```

```python
# We create a list of the names of the products from sneakersnstuff csv
names = [str(df['brand'].tolist()[i]) + " " + str(df['name'].tolist()[i]) for i in range(len(df))]

print(names)
```

```python
list_ = []
i, j = 0, 0
```

```python
coorespond = {
    "Black" : [1],
    "White" : [12, 4, 20],
    "Red" : [7, 23],
    "Grey" : [3],
    "Blue" : [9, 27, 26, 17],
    "Multi" : [15],
    "Brown" : [2],
    "Pink" : [5, 24],
    "Orange" : [11, 22, 21],
    "Yellow" : [8, 29],
    "Purple" : [6, 25],
    "Green" : [28, 30, 16, 10],
    "Gold" : [14],
    "Silver" : [13]
}
```

### Scraping Vinted

```python
list_ = []
i, k = 0, 0
nbMaxElements = 5


while i < len(names):
    try:
        start_i = i
        start_k = k
        for i in range(start_i, len(names)):
            name = names[i]
            
            params = {
                "search_text": name,
                "country":"fr",
                "catalog": [1231,16]  #the catalog is the category of the item, therefore we only want to search in the sneakers category

            }
            # We add the color to the search parameters
            if df['color'].tolist()[i] in coorespond.keys():
                params["color_ids"] = coorespond[df['color'].tolist()[i]]

            items = scraper.search(params)  # get all the items that match the search
            if(len(items) > 0):
                for k in range(start_k, len(items[:nbMaxElements])):
                    item = items[k]
                    infos = scraper.item(item.id)  # get more info about a particular item
                    new_dict = {**infos.__dict__,  'real_name' : name+" "+df['color'].tolist()[i]} # we add the name and other features of the product to a dictionary
                    list_.append(new_dict)
                    print(i + 1, "/", len(names), " ", k + 1 if len(items[:nbMaxElements]) > 0 else 0, "/", len(items[:nbMaxElements]), " ", name)
                start_k = 0
            if i == len(names) - 1:
                i += 1
    except Exception as e:
        pass
        #print(e)
```

```python
# convert list of objects to dataframe
df = pd.DataFrame.from_records(list_)
df
```

### Data preprocessing 

```python
import re
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/alexa/Downloads/vinted.csv',sep=';')

# Selecting only the required columns
new_df = df[['label','title', 'price', 'photos', 'url','user','real_name']]



# Function to extract the first photo URL from the 'photos' column
def extract_first_photo_url_v2(photos_str):
    # Regular expression pattern to find the URL
    url_pattern = r"url='(https?://\S+)'"
    match = re.search(url_pattern, photos_str)
    if match:
        return match.group(1)
    return ''

# Apply the function to the 'photos' column
new_df['photo_url'] = new_df['photos'].apply(extract_first_photo_url_v2)


```

```python
# Function to extract information from the 'user' column
def extract_info_with_regex(user_string):
    try:
        followers_count = int(re.search(r'followers_count=(\d+)', user_string).group(1))
        positive_feedback_count = int(re.search(r'positive_feedback_count=(\d+)', user_string).group(1))
        neutral_feedback_count = int(re.search(r'neutral_feedback_count=(\d+)', user_string).group(1))
        negative_feedback_count = int(re.search(r'negative_feedback_count=(\d+)', user_string).group(1))
        feedback_reputation = float(re.search(r'feedback_reputation=(\d+\.\d+)', user_string).group(1))
        feedback_count = int(re.search(r'feedback_count=(\d+)', user_string).group(1))

        return {
            'followers_count': followers_count,
            'positive_feedback_count': positive_feedback_count,
            'neutral_feedback_count': neutral_feedback_count,
            'negative_feedback_count': negative_feedback_count,
            'feedback_reputation': feedback_reputation,
            'feedback_count': feedback_count
        }
    except AttributeError:
        # If any of the attributes are not found, return None or a default value
        return None

# Apply the function to each user string in the 'user' column
new_df['user_info'] = new_df['user'].apply(extract_info_with_regex)
```

```python
def calculate_reliability_index(user_info):
    # Define a threshold for high positive feedback count
    positive_feedback_threshold = 100  # This is an example threshold, adjust as needed

    # Extract the individual metrics
    positive_feedback_count = user_info.get('positive_feedback_count', 0)
    feedback_reputation = user_info.get('feedback_reputation', 0)

    if(user_info.get('feedback_count') == 0):
        return np.nan

    # Calculate the scores
    feedback_reputation_score = feedback_reputation * 5
    positive_feedback_score = (positive_feedback_count / positive_feedback_threshold) * 5

    # Normalize the scores to not exceed 5
    positive_feedback_score = min(positive_feedback_score, 5)

    # Combine the scores with equal weight
    combined_score = (feedback_reputation_score*0.8 + positive_feedback_score*0.2) 

    # Ensure the final index is between 0 and 5
    reliability_index = min(max(combined_score, 0), 5)
    
    return reliability_index

# Apply the function to the 'user_info' column
new_df['vinted_stars'] = new_df['user_info'].apply(calculate_reliability_index)
```

```python
# Renaming the columns
new_df = new_df.rename(columns={ 'title': 'name','url':'vinted_url','price':'vinted_price','photo_url':'vinted_img'})
```

```python
new_df.drop(['photos','user','user_info','label','name'], axis=1, inplace=True)
```

```python
# Keeping the best element based on vinted_stars for each unique real_name
new_df = new_df.sort_values(by='vinted_stars', ascending=False).drop_duplicates(subset='real_name')
```
